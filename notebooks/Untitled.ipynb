```python
import geopandas as gpd
```

```python
import folium
```

```python
fay_verisi = gpd.read_file(r"C:\Users\Elfin\OneDrive\Masaüstü\proje\data\raw\DFY_GEO_WGS84_2013_s.shp")
```

```python
fay_verisi = fay_verisi.to_crs("EPSG:4326")
```

```python
print(fay_verisi.crs)
```

```python
m = folium.Map(location=[39.0, 35.0], zoom_start=6, tiles="CartoDB positron")
```

```python
folium.GeoJson(
    fay_verisi,
    name="Fay Hatları",
    style_function=lambda x: {'color': 'red', 'weight': 2, 'opacity': 0.6},
    # DİKKAT: Aşağıdaki satırda FAY_ADI yerine FAYADI yazdık
    tooltip=folium.GeoJsonTooltip(fields=['FAYADI'], aliases=['Fay Adı:']) 
).add_to(m)
```

```python
m.save("../output/turkiye_fay_haritasi.html")
```
