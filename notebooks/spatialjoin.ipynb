```python
import pandas as pd
```

```python
import geopandas as gpd
import os
```

```python
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, "..", "data")
```

```python
input_csv_path = os.path.join(data_dir, "cleaned", "kaggle_cleaned_magnitude4.csv")
input_shp_path = os.path.join(data_dir, "raw", "DFY_GEO_WGS84_2013_s.shp")
output_csv_path = os.path.join(data_dir, "processed", "depremler_ve_faylar_birlestirilmis.csv")

# Dosya yollarını kontrol etmek için yazdıralım (Hata ayıklama için)
print(f"CSV Okunacak Yer: {input_csv_path}")
print(f"SHP Okunacak Yer: {input_shp_path}")

depremler = pd.read_csv(input_csv_path)


fay_verisi = gpd.read_file(input_shp_path)
```

```python
depremler_gdf = gpd.GeoDataFrame(
    depremler,
    geometry=gpd.points_from_xy(depremler.Longitude, depremler.Latitude),
    crs="EPSG:4326"
)
```

```python
depremler_metre = depremler_gdf.to_crs("EPSG:3857")
faylar_metre = fay_verisi.to_crs("EPSG:3857")
```

```python
sonuc_df = gpd.sjoin_nearest(
    depremler_metre,
    faylar_metre,
    how="left",
    distance_col="uzaklik_metre" # Sütun adını 'uzaklik_metre' yaptık çünkü artık birim metre.
)
```

```python
sonuc_df = sonuc_df.to_crs("EPSG:4326")
```

```python
sonuc_df = sonuc_df.to_crs("EPSG:4326")

os.makedirs(os.path.dirname(output_csv_path), exist_ok=True)

sonuc_df.to_csv(
    output_csv_path,
    encoding="utf-8-sig",
    index=False
)

print(f"İşlem tamamlandı! Dosya şuraya kaydedildi: {output_csv_path}")
```
